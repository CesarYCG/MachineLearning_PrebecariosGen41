```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1, 100)
#print(x)
#(x, x al cubica)
plt.plot(x, x **3, label = "Cúbica", color="olivedrab")

#(x, x cúbica negativa)
plt.plot(x, -x ** 3, label = 'Cúbica neg', color="darkmagenta")
plt.legend(loc ="center right")#lower left,upper

plt.xlabel('eje x')
plt.ylabel('eje y')

plt.show()
```

**Colores en matplotlib**

El uso de colores se hace a través de palabras reservadas, es importante notar que para los colores más comunes existen nombres de una letra como g (green/verde), c (cyan/ciano), r (red/rojo), entre otros.
![Palabras reservadas para colores](https://drive.google.com/uc?id=1bTRpwDiq2u9UHoljjGFYngnJv_pQ_LG0)

```python
x = np.arange(0, 10, 0.2)

# Variable dependiente
y = np.sin(x)

# Creando la figura para tener donde graficar
fig = plt.figure()

fig.suptitle("Grafica") # Agrega titulo a la figura

plt.grid() # La funcion grid cuadricula el axes sobre el figure 

# Asignando los valores para graficar en plot
fig = plt.plot(x, y) 

# Asignamos nombres a los ejes
plt.xlabel('x')
plt.ylabel('Sen(x)')

# Desplazara lo contenido que esta almacenado en Figure
plt.show()
```

```python
# Variable dependiente
x=np.linspace(-1,1,50)

a=np.random.rand(50)
b =np.random.rand(50)
#print(a)


# Creando la figura para tener donde graficar
fig = plt.figure()

fig.suptitle("Diagrama de dispersion") # Agrega titulo a la figura

plt.grid() # La funcion grid cuadricula el axes sobre el figure 

# Asignando los valores para graficar en scatter, que nos imprime cada punto
fig = plt.scatter(x, b,color="r") 
fig = plt.scatter(x, a,color="b") 

# Asignamos nombres a los ejes
plt.xlabel('x') 
plt.ylabel('Sen(x)')

# Desplazara lo contenido que esta almacenado en Figure
plt.show()
```

```python
#First create some toy data:
x = np.linspace(-1, 1,10)
y = 1/x


#Crea dos axes 
fig, (ax1, ax2,ax3) = plt.subplots(3,1, sharex=True) # Retorna los Axes como tuplas
ax1.plot(x, y)
#ax1.set_title('Compartiendo eje Y')
ax2.scatter(x, y)
ax3.plot(x, y)
plt.show()
```
